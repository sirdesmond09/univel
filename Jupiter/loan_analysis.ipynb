```python
# convert applicationdate, repaymentDuedate and repaymentpaiddate in the dateframes to datetimes

data["appilcationDate"] = pd.to_datetime(data["appilcationDate"])
data["repaymentDueDate"] = pd.to_datetime(data["repaymentDueDate"])
data["repaymentPaidDate"] = pd.to_datetime(data["repaymentPaidDate"])
data.head()
```

```python
# data.info()
```

```python
interest_rate = data.InterestRate
mean_interest = round(interest_rate.mean(), 2)

term_days = data.TermDays
mean_days = round(term_days.mean(), 2)

print(mean_interest)
print(mean_days)
```

```python
data['dateDifference'] = data['repaymentPaidDate'] - data['repaymentDueDate']
data
```

```python
greater_than_3_days = data.dateDifference.dt.days > 3
late_payments = data[greater_than_3_days]
# late_payments
late = len(late_payments)
total = len(data)

```

```python
percent = (late/total)*100
print(f"{percent}%")
```
